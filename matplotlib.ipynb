```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
t = np.arange(0., 5., 0.2)

plt.figure()


#          1 row 3 cols and 1st index
plt.subplot(1, 3, 1)
plt.plot(t, t, 'r-')

#          1 row 3 cols and 2st index

plt.subplot(1, 3, 2)
plt.plot(t, t**2, 'bs')

plt.subplot(1, 3, 3)
plt.plot(t, t**3, 'g^')

plt.suptitle('Categorical Plotting')
plt.tight_layout()

plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt

# Create data
x = np.linspace(0, 10, 100)

# Create subplots
fig, axes = plt.subplots(2, 3)

# First subplot - Sin function
axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title('Sin func')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')

# Second subplot - Cos function
axes[0,1].plot(x, np.cos(x))
axes[0,1].set_title('Cos func')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')

# Third subplot - Tan function
axes[0,2].plot(x, np.tan(x))
axes[0,2].set_title('Tan func')
axes[0,2].set_xlabel('x')
axes[0,2].set_ylabel('y')

# Fourth subplot - Exponential function
axes[1,0].plot(x, np.exp(x))
axes[1,0].set_title('Exp func')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')

# Fifth subplot - Square root function
axes[1,1].plot(x, np.sqrt(x))
axes[1,1].set_title('Sqrt func')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')

# Sixth subplot - Linear function
axes[1,2].plot(x, x)
axes[1,2].set_title('Linear func')
axes[1,2].set_xlabel('x')
axes[1,2].set_ylabel('y')


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(0)
data = np.random.randn(1000)
# print(data)

# Create histogram
plt.hist(data, bins=30, color='blue', edgecolor='black')
plt.title('Simple Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
courses = ['GenAI', 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]
explode = [0.2, 0, 0, 0, 0, 0]

#autopct = auto percentage

plt.pie(students, labels=courses, shadow=True, explode=explode, autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```
